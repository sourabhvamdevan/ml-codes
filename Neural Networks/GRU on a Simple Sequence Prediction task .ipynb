```python

#Below code is an example of GRU model
#GRU stands for Gated Recurrent Unit.
#It is a type of RNN that is similar to LSTM, but with a few modifications.
#GRU has fewer parameters than LSTM, which makes it faster to train.
#GRU has two gates: reset gate and update gate.


import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
from tensorflow.keras.optimizers import Adam

#This code will generate a sine wave and then use a GRU model to predict the next value in the sequence.


def generate_data():
    X = np.linspace(0, 100, 1000)  # Sequence of 1000 values
    Y = np.sin(X)  # Sin function as target
    return X, Y

X, Y = generate_data()

#Below code will create a GRU model using Keras and train it on the generated data.
X = X.reshape((X.shape[0], 1, 1))  # 1000 samples, 1 timestep, 1 feature
Y = Y.reshape((Y.shape[0], 1))  # Corresponding output

#Below code will create a GRU model using Keras and train it on the generated data.

model = Sequential()
model.add(GRU(64, activation='tanh', input_shape=(1, 1)))  # GRU layer with 64 units
model.add(Dense(1))  # Output layer

#Below code will compile the model using the Adam optimizer and mean squared error loss function.
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

#Below code will train the model on the generated data for 200 epochs with a batch size of 64.
history = model.fit(X, Y, epochs=200, batch_size=64, verbose=2)


plt.plot(history.history['loss'])
plt.title('Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss (Mean Squared Error)')
plt.show()


predictions = model.predict(X)


plt.plot(Y, label="Actual")
plt.plot(predictions, label="Predicted")
plt.title('GRU Model - Prediction vs Actual')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()
plt.show()


print("\nModel Training Summary:")
print(f"Training Loss: {history.history['loss'][-1]:.4f}")
```
