```python


#HIERARCHICAL CLUSTERING




import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage


X, y = make_blobs(n_samples=100, centers=4, random_state=42)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


agg_clust = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_pred = agg_clust.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap='viridis', edgecolor='k', s=100)
plt.title("Agglomerative Hierarchical Clustering - 4 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


linked = linkage(X_scaled, 'ward')

plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title("Dendrogram for Agglomerative Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

```
