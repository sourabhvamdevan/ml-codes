```python

#SVD for Dimensionality Reduction

from sklearn.decomposition import TruncatedSVD
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt


digits = load_digits()
X = digits.data  


svd = TruncatedSVD(n_components=2, random_state=42)
X_reduced = svd.fit_transform(X)


print("### ORIGINAL DATASET ###")
print(f"Original shape of dataset: {X.shape} (64 features for each sample)")

print("\n### SVD DIMENSIONALITY REDUCTION ###")
print(f"Reduced dataset shape: {X_reduced.shape} (2 features for each sample)")
print("\nThe dataset has been reduced to 2 dimensions while retaining most of the information.")


plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target, cmap='viridis', s=10)
plt.colorbar(scatter, label='Digit Label')
plt.title("SVD Reduced Digits Data to 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()

print("\n### PLOT ###")
print("The scatter plot shows the reduced dataset in 2D, with clusters corresponding to digit labels.")

```
