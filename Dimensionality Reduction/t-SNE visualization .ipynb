```python
# t-SNE visualization of the handwritten digits dataset
# t-SNE(t-Distributed Stochastic Neighbor Embedding) is a popular dimensionality reduction trechnique
# that is particularly well-suited for embedding high-dimensional data into Lower-dimensional space for visualization




import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import pandas as pd


digits = load_digits()
X = digits.data  
y = digits.target  


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


print("Running t-SNE...")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=3000)
X_embedded = tsne.fit_transform(X_scaled)


df_tsne = pd.DataFrame(X_embedded, columns=['TSNE_1', 'TSNE_2'])
df_tsne['Label'] = y


plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_tsne['TSNE_1'], df_tsne['TSNE_2'], c=df_tsne['Label'], cmap='tab10', alpha=0.8, s=20)
plt.colorbar(scatter, label='Digit Label')
plt.title("t-SNE Visualization of Handwritten Digits", fontsize=16)
plt.xlabel("t-SNE Dimension 1", fontsize=12)
plt.ylabel("t-SNE Dimension 2", fontsize=12)
plt.grid(alpha=0.3)
plt.show()

#below line represents the explanation of the output
print("\n--- Explanation of the Output ---")
print("1. The plot above represents the 2D t-SNE visualization of the handwritten digits dataset.")
print("2. Each point corresponds to a digit image, reduced from 64-dimensional space to 2D space.")
print("3. Colors indicate the digit labels (0-9).")
print("4. Points with the same color (digit) form clusters, showing how t-SNE preserves local structures.")
print("5. Overlapping points may indicate similarities between digit shapes (e.g., '1' and '7').")

```
