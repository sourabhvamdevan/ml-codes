```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf
```

```python
#this will generate time series dataset.
np.random.seed(42)
n = 100  
time = np.arange(n)
data = 0.5 * np.sin(0.1 * time) + 0.2 * np.random.randn(n)
```

```python
time_series = pd.Series(data, index=time)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_series, label="Time Series Data")
plt.title("Synthetic Time Series Data")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
plot_acf(time_series, lags=20)
plt.title("Autocorrelation Function (ACF)")
plt.show()
```

```python
order = 5  # Order of the AR model (number of lags)
model = AutoReg(time_series, lags=order)
results = model.fit()
```
