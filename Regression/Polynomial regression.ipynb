```python
#polynomial Regression


import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


np.random.seed(42)
X = np.random.rand(100, 1) * 10  
y = 2 + 3 * X + 0.5 * X**2 + np.random.randn(100, 1) * 10  


plt.scatter(X, y, color="blue", label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Data")
plt.legend()
plt.show()


degree = 2  # type the degree of the polynomial you want to fit
poly_features = PolynomialFeatures(degree=degree, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Fit Polynomial Regression
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)


print("\nModel Coefficients:")
print(f"Intercept: {model.intercept_[0]:.4f}")
print(f"Coefficients: {model.coef_[0]}")

#This is for mean square error
mse = mean_squared_error(y, y_pred)
print(f"\nMean Squared Error: {mse:.4f}")

# Ploting the results
plt.scatter(X, y, color="blue", alpha=0.6, label="Data points")
plt.plot(np.sort(X, axis=0), np.sort(y_pred, axis=0), color="red", label=f"Polynomial Degree {degree}")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression Fit")
plt.legend()
plt.show()
```
