Day 1: Single Artificial Neuron using ReLU

```python
import numpy as np


X = np.array([0.8, 0.4, 1.0])


W = np.array([0.5, -0.3, 0.9])


b = 0.1


z = np.dot(W, X) + b

print("Input Vector:", X)
print("Weight Vector:", W)
print("Bias:", b)
print("Weighted Sum (z):", round(z, 4))


def relu(z):
    return max(0, z)

output = relu(z)

print("Neuron Output after ReLU:", round(output, 4))
```

Day 2: Comparison of Activation Functions


```python
import numpy as np


def relu(z):
    return max(0, z)


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def tanh_fn(z):
    return np.tanh(z)


test_values = [-3, -1, 0, 1, 3]

print(f"{'Input(z)':<10}{'ReLU':<10}{'Sigmoid':<12}{'Tanh':<12}")
print("-" * 45)

for z in test_values:
    print(f"{z:<10}{relu(z):<10.4f}{sigmoid(z):<12.4f}{tanh_fn(z):<12.4f}")
```

Day 6 : Image Generation using Variational Autoencoder (VAE)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
class VAE(nn.Module):
    def __init__(self, input_dim=784, hidden_dim=400, latent_dim=20):
        super().__init__()

        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)

        self.fc3 = nn.Linear(latent_dim, hidden_dim)
        self.fc4 = nn.Linear(hidden_dim, input_dim)

    def encode(self, x):
        h = F.relu(self.fc1(x))
        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h = F.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h))

    def forward(self, x):
        mu, logvar = self.encode(x.view(-1, 784))
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar
```

```python
def vae_loss(recon_x, x, mu, logvar):

    BCE = F.binary_cross_entropy(
        recon_x,
        x.view(-1, 784),
        reduction='sum'
    )

    KLD = -0.5 * torch.sum(
        1 + logvar - mu.pow(2) - logvar.exp()
    )

    return BCE + KLD
```

```python
transform = transforms.ToTensor()

dataset = torchvision.datasets.MNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

loader = DataLoader(
    dataset,
    batch_size=128,
    shuffle=True
)

print("Dataset Loaded Successfully!")
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("Using Device:", device)

vae = VAE().to(device)

optimizer = optim.Adam(
    vae.parameters(),
    lr=0.001
)
```

```python
epochs = 3

vae.train()

for epoch in range(epochs):

    total_loss = 0

    for images, _ in loader:

        images = images.to(device)

        recon_images, mu, logvar = vae(images)

        loss = vae_loss(
            recon_images,
            images,
            mu,
            logvar
        )

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch [{epoch+1}/{epochs}] "
        f"Loss: {total_loss:.2f}"
    )
```

```python
vae.eval()

with torch.no_grad():

    z_sample = torch.randn(16, 20).to(device)

    generated = vae.decode(z_sample).cpu()

print("Generated shape:", generated.shape)
```

```python
generated_images = generated.view(
    16,
    28,
    28
).numpy()

fig, axes = plt.subplots(
    4,
    4,
    figsize=(6, 6)
)

for i, ax in enumerate(axes.flat):

    ax.imshow(
        generated_images[i],
        cmap="gray"
    )

    ax.axis("off")

plt.show()
```

Day 8 Retrieval-Augmented Generation (RAG) and Self-Consistency Prompting using FLAN-T5

```python
!pip -q install transformers torch sentencepiece accelerate
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained(
    "google/flan-t5-large"
)

model = AutoModelForSeq2SeqLM.from_pretrained(
    "google/flan-t5-large"
)

def generate(prompt):

    inputs = tokenizer(
        prompt,
        return_tensors="pt"
    ).input_ids

    outputs = model.generate(
        inputs,
        max_new_tokens=512
    )

    return tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )

print("FLAN-T5 Loaded Successfully")
```

```python
documents = """
Document 1:
Node.js uses Google's V8 engine.

Document 2:
Express.js is a Node.js framework.

Document 3:
MongoDB is a NoSQL database.
"""

question = (
    "Which framework is commonly used with Node.js?"
)

prompt = f"""
You are a company knowledge assistant.

CONTEXT:
{documents}

RULES:

1. Answer ONLY from context.

2. If answer not found say:

Information not available in provided context.

3. Mention source document.

QUESTION:
{question}
"""

print(generate(prompt))
```

```python
from collections import Counter
import re

question = """
A company sells a product for ₹500.

Manufacturing Cost = ₹300
Marketing Cost = ₹50
Tax = ₹25

Calculate the profit per product using the following steps:

1. Calculate Total Cost

Total Cost = Manufacturing Cost + Marketing Cost + Tax

2. Calculate Profit

Profit = Selling Price - Total Cost

What is the profit per product?

Think step by step and give final answer as:

FINAL ANSWER: <value>
"""

results = []

for i in range(5):

    response = generate(question)

    print(f"\n--- Run {i+1} ---")
    print(response)

    match = re.search(
        r'(?:FINAL ANSWER|The answer)[: ]*(.*)',
        response,
        re.IGNORECASE
    )

    if match:
        results.append(
            match.group(1).strip()
        )

vote = Counter(results)

print("\n======================")
print("Majority Decision")
print("======================")

if vote:
    print(vote.most_common(1)[0])
else:
    print(
        "No valid answers were extracted."
    )
```

```python
import pandas as pd
import numpy as np
import requests
import zipfile
import io
import sqlite3
import json
import os

from datetime import datetime
from pathlib import Path

print("Ready!")
```

```python
for folder in [
    "/content/data/raw/",
    "/content/data/processed/",
    "/content/data/gold/"
]:
    os.makedirs(
        folder,
        exist_ok=True
    )

URL = (
    "https://archive.ics.uci.edu/static/public/352/online+retail.zip"
)

print("Configuration set")
```

```python
print("="*50)
print("STEP 1: Downloading Data")
print("="*50)

response = requests.get(
    URL,
    timeout=30
)

with zipfile.ZipFile(
    io.BytesIO(response.content)
) as z:

    excel_file = [
        f for f in z.namelist()
        if f.endswith('.xlsx')
    ][0]

    with z.open(excel_file) as f:

        df = pd.read_excel(
            f,
            engine='openpyxl'
        )
```

```python
df['CustomerID'] = (
    df['CustomerID']
    .fillna('UNKNOWN')
    .astype(str)
)

df['InvoiceNo'] = (
    df['InvoiceNo']
    .astype(str)
)

df['StockCode'] = (
    df['StockCode']
    .astype(str)
)

df['Description'] = (
    df['Description']
    .fillna('Unknown')
    .astype(str)
)

df['InvoiceDate'] = pd.to_datetime(
    df['InvoiceDate'],
    format='%m/%d/%Y %H:%M',
    errors='coerce'
)

df['Country'] = (
    df['Country']
    .astype(str)
)

print(
    f"Downloaded {len(df):,} rows"
)

print(
    f"Shape: {df.shape}"
)

df.head(3)
```

```python
print("="*50)
print("STEP 2: Saving Raw Data")
print("="*50)

df.to_csv(
    "/content/data/raw/raw_data.csv",
    index=False
)

print(
    "Saved: raw_data.csv"
)
```

```python
conn = sqlite3.connect(
    "/content/data/gold/retail_warehouse.db"
)

df.to_sql(
    'transactions',
    conn,
    if_exists='replace',
    index=False
)

conn.close()

print(
    "Saved to database"
)
```

```python
df.to_json(
    "/content/data/raw/raw_data.json",
    orient='records',
    lines=True
)

print("Saved as JSON")

print(
    "\nFiles saved in /content/data/raw/"
)
```

```python
print("="*50)
print("STEP 3: Adding Derived Columns")
print("="*50)

df['TotalAmount'] = (
    abs(df['Quantity'])
    * df['UnitPrice']
)

df['Year'] = (
    df['InvoiceDate']
    .dt.year
    .fillna(2010)
    .astype(int)
)

df['Month'] = (
    df['InvoiceDate']
    .dt.month
    .fillna(1)
    .astype(int)
)

df['Hour'] = (
    df['InvoiceDate']
    .dt.hour
    .fillna(0)
    .astype(int)
)

df['DayOfWeek'] = (
    df['InvoiceDate']
    .dt.dayofweek
    .fillna(0)
    .astype(int)
)

df['Weekend'] = (
    df['DayOfWeek'] >= 5
).astype(int)
```

```python
df['TransactionType'] = (
    df['InvoiceNo']
    .apply(
        lambda x:
        'CANCELLATION'
        if x.startswith('C')
        else 'SALE'
    )
)
```

```python
def get_category(code):

    first = str(code)[0] if str(code) else 'X'

    return {
        '2':'Electronics',
        '3':'Home',
        '8':'Gifts',
        '9':'Seasonal'
    }.get(first, 'Other')

df['ProductCategory'] = (
    df['StockCode']
    .apply(get_category)
)

print(
    "Derived columns created successfully"
)
```
